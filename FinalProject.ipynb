## Introduction and Background

Our final project revolves on the following question, "Are streets in San Diego with higher traffic volumes more dangerous?". We defined dangerous in two ways: number of accidents and number of injuries.

To approach this issue, we examine a dataset that contains police reports of collisions that have happened in San Diego County over the past few years, and perform various analyses on these reported collisions. These police reports give details such as time and day of the incident, what street(s) the incident took place at, and a standardized description of the incident. We also wanted to see how related car accidents and injuries were to traffic volume, so we looked at the traffic counts dataset, which is a data set that contains the number of cars that drove through any particular street section on a given date.

Our hypothesis is that streets with higher traffic volumes will experience more car accidents with more injuries, whereas the null hypothesis would be that streets with higher traffic volumes have less accidents with less injuries.

## Data Description

## Datacleaning and Preprocessing

### Collisions Dataset:

We cleaned the dataset by changing the street names to have the same street types. Some streets had "AV" and "AVENUE" to reference an avenue street type. This presented a problem when calculating the number of injuries and accidents for a specific street, so we wanted standardized the street types. In terms of wrangling the data, we wanted to get the number of accidents and injuries for every street and for every type of traffic violation. We created scripts that would find the number of accidents and injuries associated with every street name and for every type of traffic violation. We also calculated the injury per accident ratio for every street name and violation type. We also created a script that would find the top ten types of traffic violations associated with every street, so we could see why so many accidents were happening. We put all of this information into two data frames, one for streets and another for traffic violation types. The street dataframe would would contain number of accidents, injuries, and ratio for every street and the violation dataframe would contain all of that information for every violation type.

We also used Google Maps API to to Geocode our addresses so that would we could graph areas of injuries and accidents. We added the the latitude and longitude of each street to our streets dataframe as well.

### Traffic Counts Dataset:

We cleaned the dataset by changing the street types to match the street types of our collisions dataset. For example, the traffic counts dataset listed A Street as listed as "A ST" while our collisions dataset listed it as "A STREET". We wanted the full street type to be listed. We also changed streets that have numbers on them such as "01" to "01ST".  Also, some of the traffic counts for specific directions were listed as NaN. The NaN's indicated that no traffic was counted on those roads. We changed all of those NaN's to be listed as zero. 

In terms of wrangling the data, we wanted to get the total number of traffic counts per road. We found all of the unique roads and wrote a script that would sum up all of the traffic counts associated with each street. We then created a dataframe that listed the total traffic count for every street. We merged this dataframe with our street dataframe that contained information on the number of accidents and injuries for every street.
```python
charges = df['charge_desc'].value_counts()
streets = df['street_name'].value_counts()
full_streets = df['full_street_name'].value_counts()
```

```python
def get_top_accidents(street):
    return list(df[df['full_street_name'] == street]['charge_desc'].value_counts().keys())[:10]
def get_top_streets(accident):
    return list(df[df['charge_desc'] == accident]['full_street_name'].value_counts().keys())[:10]
```

#### We are cleaning the traffic counts dataset

```python
first, *middle, last = traffic_counts_df['street_name'][0].split()
street_types = set()
for street in traffic_counts_df['street_name']:
    if len(street.split()) > 1:
        first, *middle, last = street.split()
    street_types.add(last)
```

```python
street_transfer = {'ST': 'STREET', 'AV':'AVENUE', 'BL': 'BOULEVARD', 'DR': 'DRIVE', 'W':'WAY', 'WY': 'WAY', 'PKWY':'PARKWAY','PKY':'PARKWAY', 'PY':'PARKWAY', 'PZ':'PLAZA', 'PL':'PLACE', 'CT':'COURT', 'LN':'LANE', 'HWY':'HIGHWAY', 'HY':'HIGHWAY', 'CR':'CIRCLE', 'ML':'MALL', 'TR': 'TERRACE', 'RD':'ROAD'}

def change_type(street):    
    if len(street.split()) > 1:
        first, *middle, last = street.split()
        if last in street_transfer:
            last = street_transfer[last]
        middle.insert(0, first)
        middle.append(last)
        new_street = ' '.join(middle)
        return (new_street + ', SAN DEIGO')
    
    return (street + ', SAN DEIGO')

def change_num(street):
    if len(street.split()) > 1:
        first, *middle, last = street.split()
        if len(first) == 2 and first[0:1].isdigit() and first[1:2].isdigit():
            if first[1] == '0':
                first = first + "TH"
            elif first[0] == '1':
                first = first + "TH"
            elif first[1] == '1':
                first = first + "ST"
            elif first[1] == '2':
                first = first + "ND"
            elif first[1] == '3':
                first = first + 'RD'
            else:
                first = first + "TH"
            
        middle.insert(0, first)
        middle.append(last)
        new_street = ' '.join(middle)
        return new_street
    
    return street
```

```python
traffic_counts_df['street_name'] = traffic_counts_df['street_name'].apply(change_type)
traffic_counts_df['street_name'] = traffic_counts_df['street_name'].apply(change_num)
street_array = traffic_counts_df['street_name'].unique()
```

#### We are summing up the total traffic for every street

```python
street_total_traffic = []
for street in street_array:
    street_total_traffic.append(sum(list(traffic_counts_df[traffic_counts_df['street_name'] == street]['total_count'])))
```

```python
traffic_volume = pd.DataFrame({'street_name': street_array, 'total_traffic': street_total_traffic}, columns=['street_name', 'total_traffic'])
traffic_volume
```

### Hourly Analysis

After performing analyses on number of incidents to number of people injured, as well as observing frequency of charges, analyzing time of day with relation to collisions would be an important perspective to analyze as well.

```python
mySeries = mySeries[mySeries != '00:01:00']
```

```python
# Function to convert hour:minute:second format to purely seconds, then return as a float less than 24
def convert_time(str_in):
    h, m, s = (str(str_in)).split(':')
    result = int(h) * 3600 + int(m) * 60 + int(s)
    return float(result / 3600)
```

One interesting note is that according to the dataset, there are significantly more incidents that occur at 12:01am more than any other time of day. This is most likely what officers used as a placeholder for times they were unable to report accurately. So we will be removing that time since we do not know how many of these incidents are genuine collisions reported at that exact time. 

```python
time_coordinates = mySeries.apply(convert_time).value_counts()
```

```python
# Generate scatter plot of time of day over frequency of collision at that time

time_x = list(time_coordinates.keys())
time_y = list(time_coordinates)
plt.scatter(time_x, time_y)
plt.xlabel('Time of day (in hours)')
plt.ylabel('Number of collisions')
plt.show()
```

From the visual above, it's awfully clumped together near the bottom but we can barely make out a bit of a trend between what appears to be 12 and 18, which represents 12:00pm to 6:00pm, which sounds like rush hour. To make this more apparent, we will remove clumping, as these points are plotted by the second, and instead round them to every half hour.

```python
# Function to convert hour:minute:second format to purely seconds, then return as a float less than 24 rounded by .5
def convert_time_estimate(str_in):
    h, m, s = (str(str_in)).split(':')
    if (int(m) > 45):
        resultEstimate = (int(h) + 1) * 3600
    elif (int(m) > 15 and int(m) < 45):
        resultEstimate = int(h) * 3600 + 30 * 60
    else:
        resultEstimate = int(h) * 3600
    return float(resultEstimate / 3600)
```

```python
time_coordinates_estimate = mySeries.apply(convert_time_estimate).value_counts()
```

```python
# Generate graph based on 30 minute intervals

estimate_x = list(time_coordinates_estimate.keys())
estimate_y = list(time_coordinates_estimate)
plt.scatter(estimate_x, estimate_y)
plt.xlabel('Time of day (in hours)')
plt.ylabel('Number of collisions')
plt.show()
```

Approximating the number of collisions to thirty minute intervals makes the data much easier to interpret. The trend between 12:00pm and 6:00pm is even more apparent, suggesting that rush hour really brings out the worst in drivers. The interval between 12:00am and 3:00am is still notable and might suggest incidents of sleeping behind the wheel, DUIs, or poor lighting of streets as well.
